```python
import matplotlib.pyplot as plt
import networkx as nx

# get graph - pre-established in networkx
G = nx.krackhardt_kite_graph()

# set the layout of the graph
pos = nx.spring_layout(G, scale=7)

# set attributes for nodes
nx.draw_networkx_nodes(G, pos,
                       nodelist=[0, 1, 2, 3, 4],
                       node_color='r',
                       node_size=1000,
                       alpha=0.5)

nx.draw_networkx_nodes(G, pos,
                       nodelist=[5, 6, 7, 8, 9],
                       node_color='b',
                       node_size=1000,
                       alpha=0.5)

# some labels for nodes
labels = {}
labels[0] = r'$Andre$'
labels[1] = r'$Beverly$'
labels[2] = r'$Carol$'
labels[3] = r'$Diane$'
labels[4] = r'$Ed$'
labels[5] = r'$Fernando$'
labels[6] = r'$Garth$'
labels[7] = r'$Heather$'
labels[8] = r'$Ike$'
labels[9] = r'$Jane$'

#plot graph
nx.draw(G, pos)
nx.draw_networkx_labels(G, pos, labels, font_size=16)
plt.show()
```
